```python
import xlrd
import datetime as dt

data['InvoiceDate'] = pd.TimedeltaIndex(data['InvoiceDate'], unit='d') + dt.datetime(1899, 12, 30)
```

```python
data['CustomerID'] = data['CustomerID'].astype(int)
```

```python
data = data[data['Quantity'] >= 0]
```

```python
data.describe()
```

```python
data.head()
```

```python
data['Price'] = data['UnitPrice'] * data['Quantity']
```

```python
data.head()
```

```python
needed_data = data[['CustomerID', 'InvoiceDate', 'Price', 'InvoiceNo']]
```

```python
grouped_data = needed_data.groupby('CustomerID').agg(Monetary = ('Price', np.sum), Max_date = ('InvoiceDate', max), Frequency = ('InvoiceNo', np.unique)).reset_index()
```

```python
grouped_data
```

```python
curr_date = pd.to_datetime('today').date()
grouped_data['Current_date'] = curr_date
```

```python
grouped_data
```

```python
grouped_data['Recency'] = (pd.to_datetime(grouped_data['Current_date']) - pd.to_datetime(grouped_data['Max_date'])).dt.days
```

```python
grouped_data
```

```python
grouped_data['Frequency'] = grouped_data['Frequency'].apply(lambda x: len(x))
```

```python
grouped_data
```

```python
final_data = grouped_data[['CustomerID', 'Recency', 'Frequency', 'Monetary']]
```

```python
final_data
```

```python
final_data['R'] = final_data['Recency'].rank(ascending = False)
final_data['F'] = final_data['Frequency'].rank(ascending = True)
final_data['M'] = final_data['Monetary'].rank(ascending = True)
```

```python
"""final_data['R'] = (final_data['R'] * 100) / final_data['R'].max()
final_data['F'] = (final_data['F'] * 100) / final_data['F'].max()
final_data['M'] = (final_data['M'] * 100) / final_data['M'].max()"""
```

```python
final_data
```

```python
final_data.to_excel('RMF_analysis_final.xlsx')
```

```python
from sklearn.cluster import KMeans
```

```python
X = final_data[['R', 'F', 'M']]
```

```python
SSE=[]
for i in range(1,10):
    KM=KMeans(n_clusters=i,max_iter=500)
    KM.fit(X)
    # APPEND THE SSE
    SSE.append(KM.inertia_)

# plot the graph -- elbow approach  x axis- no. of cluster, Y axis-- SSE

plt.figure(figsize=(10,5))
plt.plot(range(1,10),SSE,color="b",marker="o")
plt.xlabel("no. of clusters")
plt.ylabel("sum of square error")
```

```python
km = KMeans(n_clusters=4)
km.fit(X)
y=km.predict(X)
final_data['Cluster'] = y
```

```python
final_data
```

```python

average_rfm = final_data.groupby('Cluster').agg(R_mean = ('R', np.mean), F_mean = ('F', np.mean), M_mean = ('M', np.mean)).reset_index()
```

```python
average_rfm
```

```python
def segmentation(cluster):
  if cluster == 0:
    return 'Moderate customer'
  elif cluster == 1:
    return 'Inactive customer'
  elif cluster == 2:
    return 'Churn risk'
  else:
    return 'High valued customer'

final_data['Segment'] = final_data['Cluster'].apply(segmentation)
```
